Uzupełnij kod tak, aby wydrukować tylko 3 pierwsze elementy z listy *pelnalista*

### Zadanie 2
Napisz 2 pętle: 

* Pierwsza doda do pustej listy wszystkie kolejne wielkorotności 3 - od 3 do 21
* Druga doda te same liczby w odwrotnej kolejności 

```python
pustaLista = []

## tu jest miesce na wpisanie pętli

#Pętla 1 - zwróccie uwagę, że aby dodać 21 konieczne jest zakończenie pętli powyżej tej wartości 
for i in range(3, 22, 3):
  pustaLista.append(i)

#Pętla 2 - w tym przypadku mamy ujemny krok - pętla powinna zakończyć się poniżej 3 np. na 2
for i in range(21, 2, -3):
  pustaLista.append(i)

## koniec pętli
    
print(pustaLista)
```

### Zadanie 4
Użyj pętli zagnieżdzonej, aby wydrukować wszystkie elementy w liście list

```python
listaList = [["Juventus", "Inter", "AC Milan"],
             ["Real Madryt", "FC Barcelona", "Atletico Madryt "],
             ["Manchester United", "Tottenham Hotspurs", "Leicester City"]]

# W pierwszej pętli iterujemy po dostępnych listach 
for lista in listaList:
  #Wydrukujmy sobie nawet listę aby zobaczyć gdzie jesteśmy
  print(lista) 
  
  #W drugiej iterujemy po konkretnych elementach wybranej listy
  for element in lista: 
    print(element)
```

### Zadanie 5

Użyj pętli zagnieżdzonych aby przekształcić listę królów polski, w takiej kolejności aby znaki były od tyłu do przodu.

**Popdowiedzi**:
* W pętli którą przeglądasz listę każdorazowo twórz pusty tekst string ("")
* Następnie wykonuj pętle opartą na indeksach, która przejdzie od ostatniego znaku w tekście do pierwszego. Znaki dodawaj do pustego tekstu (konkatenacja) Uprościsz sobie pracę wykorzystując ujemne indeksowanie.
* Po przejsciu pętli dodawaj wygenerowany tekst do listy *krolowiePolskiOdtylu*. 


```python
krolowiePolski = ["Przemysł II", "Władysław Łokietek", "Kazimierz Wielki"]

krolowiePolskiOdtylu = []

## tu jest miesce na wpisanie pętli

# W pętli iterujemy po elementach listy
for element in krolowiePolski:
  #Zaczynamy od stworzenia pustej zmiennej, która zawierać będzie imię króla prezentowane od tyłu
  NazwaOdTylu = ""

  # W drugiej pętli będziemy iterować po kolejnych znakach tekstu.
  # Aby to zrobić wykorzystamy obiekt range i indeksy od pierwszego do długości tekstu.
  # Do końca zakresu trzeba dodać 1 stąd zapis (len(element) + 1)
  for i in range(1, len(element)+1):
    #Będziemy dodawać kolejne wyrazy przy wykorzystaniu ujemnego indeksowania 
    NazwaOdTylu = NazwaOdTylu + element[-i] 

  # Na końcu dodajemy nazwę do finalnej pętli
  krolowiePolskiOdtylu.append(NazwaOdTylu)
## koniec pętli

print(krolowiePolskiOdtylu)
```
