## Pakiet NumPy - obiekt NDarray
NumPy wprowadza do języka Python dodatkowy typ danych ndarray - pozwala on tworzyć macierze wielowymiarowe. 

Macierze te w składni przypominać będą trochę listy list, ale ich funkcjonalność jest całkiem inna. 

Przykładowy wydruk takiej macierzy poniżej:





```python
import numpy as np

a = np.arange(15).reshape(3, 5)

# Wydrukujemy element 
print(a)
print("")

# Oraz informacje o jego typie:
print("Typ tablicy:", type(a))
```

Tablice języka NumPy posiadają kilka funkcji diagnostycznych - obrazują one:
* *shape* - rozmiar,
* *ndim* - liczbę wymiarów
* *size* - łączną liczbę elementów
* *dtype* - typ elementu - w tej bibliotece jest to o tyle ważne, że nie istnieje np. macierzowe łączenie tablic typu int i float

Wydrukujmy te parametry dla naszej tablicy a:

```python
print("Nasza tablica ma następujący rozmiar:", a.shape)
print("Oznacza to, że ma:", a.ndim ,"wymiary")
print("i składa się z :", a.size ,"elementów typu", a.dtype)
```

### Tworzenie macierzy 
Macierze powstają na podstawie list bądź krotek. W przypadku elementów wielowymiarowych będą to listy krotek albo listy list. 

```python
print("Tablica jednowymiarowa wygląda tak:")
Tablica1D = np.array([1, 2, 3, 4])
print(Tablica1D)

print()
print("Tablica dwuwymiarowa wygląda tak:")
Tablica2D = np.array([(1, 2), (3, 4)])
print(Tablica2D)

print()
print("Tablica trójwymiarowa liczb zespolonych wygląda tak:")
Tablica3D = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]], dtype=complex)
print(Tablica3D)
```

Istnieje kilka metod wartych zapamiętania do tworzenia różnych macierzy:

* Zeros - tworzy macierz samych zer: 

```python
a = np.zeros((3,3))
print(a)
```

* Ones - tworzy macierz samych jedynek:

```python
a = np.ones((3,3))
print(a)
```

* Metoda *eye* - tworzy jednostkową macierz diagonalną. 
* Kombinacja metod *diag* i *full* - Metoda *full* tworzy macierz o podanych rozmiarze i wpisuje do niej wartość z drugiego argumentu. Metoda *diag* wykorzystuje to do stworzenia przekątnej. 

```python
print("Metoda eye:")
a = np.eye(2)
print(a)

print()
print("Metoda full:")
b = np.full(5,3)
print(b)

print()
print("Metoda diag:")
a = np.diag(b) 
print(a)
```

* *arange* - umożliwia tworzenie ciągów na podstawie zbliżonej do obiektu *range* (tego od pętli *for*). Zwykle używana w połączeniu z metodą *reshape*.

```python
print("Ciąg zrobiony metodą arange:")
# Argumenty to kolejno początek, koniec i krok
a = np.arange(3, 45, 3)
print(a, "\n")

print("Po przerobieniu metodą reshape:")
b = a.reshape(7,2)
print(b)

```

* *linspace* - dzieli odcinek na x elementów. Kolejne argumenty funkcji oznaczają: początek odcinka, koniec odcinka oraz ilość elementów na które ma zostać podzielony odcinek. 

```python
a = np.linspace(1,100, 20)
print(a)
```

* *fromfunction* - tworzy tablicę na podstawie funkcji.

```python
def f(x, y):
    return 10 * x + y

b = np.fromfunction(f, (5, 4), dtype=int)
print(b)
```

### Indeksowanie przy macierzach
Indeksowanie odbywa się podobnie do list.

Pewnym nowum jest operator :, który oznacza, aby wybrać cały wiersz / kolumnę. 

W blokach kodu poniżej przedstawiony zostanie indeksowanie w tabeli dwuelemntowej: 

```python
a = np.array([(1,1), (2,2), (3,3), (4,4)])
# pojedyczny indeks zwraca wiersz: 
a[1]
```

```python
a = np.array([(1,1,11), (2,2,22), (3,3,33), (4,4,44)])
# Aby wybrać określony element odnosimy się dwoma argumentami: 
a[1,2]
```

```python
a = np.array([(1,1,11), (2,2,22), (3,3,33), (4,4,44)])
# Taki kod wybierze nam trzecią kolumnę: 
a[1:3,2]
```

```python
a = np.array([(1,1,11,111), (2,2,22,222), (3,3,33,333), (4,4,44,444)])
# Taki kod wybierze nam trzeci wiersz: 
a[2,:]
```

Uwaga: działa również ujemne indeksowanie:

```python
a = np.array([(1,1), (2,2), (3,3), (4,4)])
# ujemny indeks zwraca ostatni wiersz: 
a[-1]
```

### Pętle na macierzach

Macierze podobnie jak słowniki posiadają kolekcje po których można iterować przy pomocy pęti. 

Standardowa pętla będzie przeglądać kolejne wiersze:

```python
a = np.array([(1,1), (2,2), (3,3), (4,4)])
for row in a:
  print(row)
```

Aby operować bezpośrednio na elementach wykorzystujemy kolekcję *flat* - najpierw przeglądane są wszystkie wartości w konrketnym wierszu, potem następuje przeskok do kolejnego wiersza.

```python
a = np.array([(1,1), (2,2), (3,3), (4,4)])
for element in a.flat:
  print(element)
```

### Operacje na macierzach
Podstawowe operacje arytmetyczne na elementach macierzach wykonywane są przez zwykłe mnożenia, dodawania, potęgowania etc. 

Każdy element pierwszej tablicy operuje z odpowiadającym mu elementem drugiej tablicy:

```python
a = np.array([(1,2),(2,1)])
b = np.array([(2,3),(3,2)])

print("Macierz A:")
print(a, "\n")

print("Macierz B:")
print(b, "\n")

print("Macierz C = A + B")
c = a + b 
print(c, "\n")

print("Macierz D = elementowo A * B")
d = a * b 
print(d, "\n")

print("Macierz E = elementowo A do potęgi B")
e = a ** b 
print(e, "\n")
```

Prawdziwym benefitem jest jednak możliwość operowania macierzowego. Podstawowe działania omówione są poniżej: 

* Transpozycja - obraca tablicę tak, że wiersze stają się kolumnami i vice versa. 

```python
a = np.array([(1,2,3), (2,3,4)])
b = a.T

print("Początkowa tablica:")
print(a)
print()
print("Tablica po transpozycji:")
print(b)
```

* Mnożenie macierzy - do obliczeń wykorzystany jest nowy operator @ albo funkcja dot.

```python
a = np.array([(1,2),(2,1)])
b = np.array([2,2])

c = a @ b 
print(c)

d = a.dot(b)
print(d)
```

* odwrócenie macierzy czy liczenie wyznaczników:

```python
a = np.array([(1,2),(2,1)])
b = np.linalg.inv(a)
print(b)
```

### Konwersje danych - Pandas i NumPy
Pakiety *Pandas* i *NumPy* mogą łatwo transferować dane między sobą.

W przypadku *Pandas* robi to metoda *to_numpy()*

```python
import numpy as np
import pandas as pd

dictToPandas = {"liczby1": [0,1,2,3], "liczby2": [1,2,3,4]}

df = pd.DataFrame(dictToPandas)
print("Wydruk DataFrame:")
print(df, "\n")

print("Wydruk Numpy:")
a = df.to_numpy()
print(a)
```

W przypadku transferu w drugą stronę - działa to podobnie jak przy liście krotek. 

Trzeba określić nazwy kolumn i wierszy:

```python
numpy_data = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(data=numpy_data, index=["row1", "row2"], columns=["column1", "column2"])
print(df)
```

## NumPy - przykłady

Za pomocą NumPy stosunkowo łatwo obrazować wykresy gęstości - tutaj przyklad dla geograficznej odległości: 

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
xs, ys = np.meshgrid(x, y, sparse=True)

# Liczymy wartości funkcji jako kwadrat współrzędnych x i y
zs = np.sqrt(xs**2 + ys**2)

# Wykres konturowy
h = plt.contourf(x, y, zs)
plt.axis('scaled')
plt.colorbar()
plt.show()
```

Przykład z dokumentacji Numpy - Fraktal Mandelbrota:

```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20, r=2):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much

    return divtime
plt.imshow(mandelbrot(400, 400))
```
