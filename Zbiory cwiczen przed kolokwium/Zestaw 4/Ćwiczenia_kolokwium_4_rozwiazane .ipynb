## Zadanie 1 - Pętla for
Napisz program, który pobiera od użytkownika liczbę całkowitą. Następnie liczy wartości następujących wyrażeń:
1. Sumę kwadratów i sześcianów kolejnych liczb całkowitych od 1 do podanej wartości. Matematycznie przedstawia to formuła:
$$  \sum_{k=1}^n (x^{2} + x^{3}) $$

2. Iloraz pierwiastków kolejnych liczb całkowitych od 1 do podanej wartości. Formuła: 
$$  \prod_{k=1}^n \sqrt{x} $$

**Podpowiedź:** Pierwiastek można wykonać prostym operatorem podwójnego mnożenia - to podniesienie do potęgi równej $\frac{1}{2} $


````python
# Wczytanie liczby
liczba = input("Podaj liczbę:")
liczba = int(liczba)

# Zainicjowanie sumy iloczynu 
suma = 0
iloczyn = 1

# Pętla
for i in range(1,liczba + 1):
  suma += i**2 + i**3
  iloczyn *= i**(1/2)

# Wydruk 
print("Suma:", suma)
print("Iloczyn:", iloczyn)
````

## Zadanie 4 - Metody string i funkcja map.
Napisz wyrażenie lambda bądź funkcję, które: 
1. Usunie wszystkie niepotrzebne białe znaki z tekstu.

3. Powiększy każdą z liter. 

Następnie wykorzystaj funkcję *map*, aby zaaplikować ją do listy

````python
listaFootball = ["  Chelsea  ", "  Liverpool  ", "  Leicester City  "]

# Dłuższe rozwiazanie
def funkcjaTekst(tekst):
  tekst = tekst.strip() # Usuwam białe znaki
  tekst = tekst[0:3] # pierwsze 3 -elementy 
  tekst = tekst.upper() # Powiększenie liter
  return tekst

listaWynikowa = list(map(funkcjaTekst, listaFootball))
print(listaWynikowa)

# Krótsze
f = lambda x: x.strip()[0:3].upper()
listaWynikowa2 = list(map(f, listaFootball))

print(listaWynikowa2)
````

## Zadanie 5 - Funkcje
Napisz program, który policzy średnie: arytmetyczną, harmoniczną i geometryczną na podstawie elementów listy. Następnie zwróci te wartości w formie krotki. 

Dla przypomnienia formuły kolejnych średnich: 

$$ średnia\ arytmetyczna = \frac{x_1 + x_2 + .... + x_{n-1} + x_n }{n} $$
$$ $$
$$ średnia\ harmoniczna = \frac{n}{\frac{1}{x_1} + \frac{1}{x_2} + .... + \frac{1}{x_{n-1}} +\frac{1}{x_{n}}} $$
$$ $$
$$ średnia\ geometryczna = \sqrt[n]{x_1*x_2* .... * x_{n-1}*x_m} $$

````python
TakieOcenyMacieOtrzymac = [5,5,5, 4.5, 4.5, 4]

def funkcjaSrednie(listaElementow):
  # Ustawiam wszystkie sumy na zero, iloczyn na 1 - nazwa oznacza przeznaczenie zmiennej
  srednia_arytmetyczna_licznik = 0
  srednia_harmoniczna_mianownik = 0
  srednia_geometryczna_podPierwiastkiem = 1
  n = len(listaElementow) # ilość elementów

  # Zliczam sumy ze wszystki elementów każdej listy
  for element in listaElementow:
    srednia_arytmetyczna_licznik += element
    srednia_harmoniczna_mianownik += 1/ element
    srednia_geometryczna_podPierwiastkiem *= element

  # Liczę właściwe średnie:
  srednia_arytmetyczna = srednia_arytmetyczna_licznik / n
  srednia_harmoniczna = n / srednia_harmoniczna_mianownik
  srednia_geometryczna = srednia_geometryczna_podPierwiastkiem**(1/n)

  return (srednia_arytmetyczna, srednia_harmoniczna, srednia_geometryczna)

funkcjaSrednie(TakieOcenyMacieOtrzymac)
````

## Zadanie 6 - Funkcje
Napisz funkcję, która policzy wartośc wariancji liczb na podstawie następujących definicji: 
1. Kwadrat odchyleń poszczególnych elementów od średniej: 
$$ Var(x) = \frac{1}{n} * \sum_{i=1}^n (x_i - \overline{x})^2 $$
2. Różnica średniej z wartości podniesionych do kwadratu oraz kwadratu standardowej średniej: 
$$ Var(x) = \mathbf{E}(X^2) - \mathbf{E}(X)^2 $$

W obu przypadkach pomoże funkcja wyprowadzona w zadaniu 5.

````python
TakieOcenyMacieOtrzymac = [5,5,5, 4.5, 4.5, 4]

def funkcjaWariancje1(lista):
  # Informacje o średniej elementów pobieram z funkcji z zadania 5
  srednia_arytmetyczna = funkcjaSrednie(lista)[0]

  # Liczę odchylenie każdego elementu od sredniej
  listaOdchylenOdSredniej = [element - srednia_arytmetyczna for element in lista]
  
  # Podnoszę te wartości do kwadratów
  listaKwadratowOdchylenOdSredniej = [element**2 for element in listaOdchylenOdSredniej]

  # Sumuje kwadraty odchyleń od średniej
  suma = 0 
  for element in listaKwadratowOdchylenOdSredniej:
    suma += element

  # Dzielę te wartości przez liczbę elementów - otrzymuje wariancję
  return suma / len(listaKwadratowOdchylenOdSredniej) 

print(funkcjaWariancje1(TakieOcenyMacieOtrzymac))
````

````python
TakieOcenyMacieOtrzymac = [5,5,5, 4.5, 4.5, 4]

def funkcjaWariancje2(lista):
  # Tworzę listę zmiennych podniesionych do kwadratu - to pierwszy człon definicji
  listaDoKwadratu = [element ** 2 for element in lista]
  
  # Zliczam sumy wszystkich elementów listy oraz elemetnów podniesionych do kwadratów
  srednia_kwadraty = 0
  srednia_listy = 0
  for i in range(0, len(lista)):
    srednia_kwadraty += listaDoKwadratu[i]
    srednia_listy += lista[i]
  
  # Przeszktałcam sumę w średnią 
  srednia_kwadraty /= len(lista)
  srednia_listy /= len(lista)

  # Zwracam wariancję zgodnie ze wzorem
  return srednia_kwadraty - srednia_listy**2

print(funkcjaWariancje2(TakieOcenyMacieOtrzymac))
````

## Zadanie 7 - Funkcje i rekurencja
Napisz program, który wygeneruje sumę ciągu Fibonacciego.

Funkcja powinna zwracać: 
* 0 dla argumentu 0
* 1 dla argumentu 1
* Sumę własnych wywołań dla wartości n-1 i n-2 dla każdej innej liczby.

Matematycznie wygląda to tak: 

$$ F(x) = \begin{cases}
0 \quad dla\ x = 0 \\
1 \quad dla\ x = 1 \\
F(x-2) + F(x-1) \quad dla \ x \ge 2  
\end{cases} $$


````python
def funkcjaFibo(wartosc):
  if wartosc == 0: 
    return 0
  elif wartosc == 1:
    return 1
  else:
    return funkcjaFibo(wartosc - 2) + funkcjaFibo(wartosc - 1) 
  
i = input("Wybierz liczbe:")
i = int(i)

print("Wynik funkcji:", funkcjaFibo(i))
````

## Zadanie 10 - Listy, zbiory i słowniki 
Wykonaj następujące polecenie: 
1. Wydrukuj lata w których Real Madryt zdobywał ligę mistrzów.
2. Wypisz nazwy klubuów, które sięgały po trofeum. 
3. Zlicz liczbę triumfów każdego zespołu. 

````python
finalisciLM = [("FC Barcelona",2011), ("Chelsea",2012), ("Bayern Monachium",2013), ("Real Madryt",2014), ("FC Barcelona",2015), 
               ("Real Madryt",2016), ("Real Madryt",2017), ("Real Madryt",2018), ("Liverpool",2019), ("Bayern Monachium",2020),
               ("Chelsea", 2021)]

# Punkt 1:
# Wyniki pozyskuje w pętli.
punkt1 = [element[1] for element in finalisciLM if element[0]== "Real Madryt"]
print("Zwycięstwa Realu: ", *punkt1)

# Punkt 2:
# Jeżeli chcemy uzyskać listę unikatów - wybieramy set. 
# W pierwszej klejności chcemy wszystkie nazwy drużyn - robimy to w pętli:
druzyny = [element[0] for element in finalisciLM]

# Następnie robimy prostą konwersję i drukujemy wynik
punkt2 = set(druzyny)
print("Zwyciescy LM:", punkt2)

# Punkt 3:
# Do liczenia ilości wystąpień najlepiej wykorzystać słownik. 
triumfy = {}
for element in finalisciLM:
  # Jeżeli nie ma elementów w słowniku tworzy go:
  if element[0] not in triumfy:
    triumfy[element[0]] = 1 
  # W innym przypadku dopisuje +1 do dotychczasowej liczby wystąpień. 
  else: 
    triumfy[element[0]] += 1

print("Liczba triumfów:" triumfy)
````
