## Zadanie 1 - warunek if
Napisz program, który otrzyma od użytkownika liczbę całkowitą a w bloku *input*. Następnie sprawdzi co jest większe: $ a^{e}$ czy ${e}^a$.

Podpowiedź: Wartość ${e}$ zapisana jest jako stała w bibliotece Math:

```python
# Import math Library
import math

# Print the value of E
print (math.e)
```

Zadanie rozwiąż w bloku poniżej: 

```python
a = int(input("Wpisz liczbę:"))

e = math.e

if a ** e > e ** a: 
  print("Wiesze jest a do potegi e")
elif a ** e == e ** a:
  print("Obie są równe")
else:
  print("Wiesze jest e do potegi a")
```

## Zadanie 2 - pętla while
Napisz program, który otrzymuje od użytkownika liczbę całkowitą a. Następnie tworzy dwie listy: 
* Pierwsza zawierać będzie kolejne liczby począwsze od 1 podniesione do potęgi a
* Druga zawierać będzie wartość a podniesioną do kolejnych potęg.

Elementy dodawaj tak dopóki, dopóty różnica między kolejnymi elementami obu list nie przekroczy 100

```python
a = int(input("Wpisz liczbę:"))
i = 1 

lista1 = []
lista2 = []

while abs(a**i - i**a) < 100: 
  lista1.append(a**i)
  lista2.append(i**a)
  i += 1 

print("Funkcja kolejne podstawy:", lista1)
print("Funkcja kolejne wykładniki:", lista2)
```

## Zadanie 3 - listy
Poniższy blok wygeneruje losową listę liter - uruchom go: 

```python
import random
import string

losowa_listaLiter = []
for i in range(0,100):
  n = random.choice(string.ascii_lowercase)
  losowa_listaLiter.append(n)

print(losowa_listaLiter)
```



Następnie napisz program, który usunie z listy wszystkie wartości 'a'.

**Podpowiedź**: Zadanie można zrobić w jednej linijce kodu

```python
# Rozwiazanie 1 - pętla while
rozwiazanie1 =  losowa_listaLiter.copy() 
while 'a' in rozwiazanie1:
  rozwiazanie1.remove('a')

# Rozwiazanie 2 - petla for
rozwiazanie2 = [element for element in losowa_listaLiter if element is not 'a' ]

# Zobaczmy ile elementów usunęło
print(len(rozwiazanie2))

# Test czy jest litera A
'a' in rozwiazanie2
```

## Zadanie 4 - zbiory 
Stwórz kolejne zmienne: 
* zbior2 - będący kopią zbior1 z dodaną wartością 999 
* zbior3 - będący suma zbiorow zbior1 i zbior_do_dodania
* zbior4 - zawierający tylko te wartości ze zbior1 i zbior_do_dodania, które występują tylko w jednym ze zbiorów. 

**Podpowiedzi:**
* Każdą bardziej złożoną metodę w Python można kopiować metodą *copy*()
* W zbiorach inaczej dodaje się pojedynczą liczbę, a inaczej inną strukturę. 

```python
zbior1 = {1,2, 100}
zbior_do_dodania = {100,200, 300}

zbior2 = zbior1.copy()
zbior2.add(999)
print(zbior2)

zbior3 = zbior1.union(zbior_do_dodania)
print(zbior3)

zbior4 = zbior1.symmetric_difference(zbior_do_dodania)
print(zbior4)
```
