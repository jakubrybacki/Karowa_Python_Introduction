## Zadanie 1
Stwórz listę liczb od 1 do 5 - bedzie to nasza bazowa lista. Następnie przygotuj: 

* drugą listę, która weźmie pierwsze trzy elementy z listy bazowej 
* trzecią listę, w której trzeci element listy bazowej zostanie zmieniony na 100
* czwartą listą, gdzie lista bazowa zostanie rozszerzona o dodatkowe 2 elementy wpisane za pomocą listy. 
* piątą listę, w której z listy bazowej usunięta zostanie wartośc 4.  

Wpisz kody które wydrukują każdą listę jedną po drugiej. 

```python
# lista 1
lista1 = [1,2,3,4,5]

# lista 2
lista2 = lista1[0:3]

# lista 3
lista3 = lista1.copy()
lista3[2] = 100

# lista 4
lista4 = lista1.copy()
lista4.extend([6,7])

# lista 5
lista5 = lista1.copy()
lista5.remove(4)

# Wydruki
print("Lista nr. 1:", lista1)
print("Lista nr. 2:", lista2)
print("Lista nr. 3:", lista3)
print("Lista nr. 4:", lista4)
print("Lista nr. 5:", lista5)
```

## Zadanie 2: Instrukcje If, elif, else
Do napisania jest następujący program -  użytkownik podaje numer poprzez komendę *input*. 

Następnie program wyświetli jedną z poniższych informacji, czy liczba jest podzielna:
* Zarówno przez 3 i przez 5
* Tylko przez 3
* Tylko przez 5
* Nie jest podzielna przez żadną z tych liczb

```python
a = input("Wpisz liczbę całkowitą:")
a = int(a)

if a%3 == 0 and a%5 == 0:
  print("Liczba jest podzielna przez 3 i 5")
elif a%3 == 0:
  print("Liczba jest podzielna wyłącznie przez 3")
elif a%5 == 0:
  print("Liczba jest podzielna wyłącznie przez 5")
else:
  print("Liczba nie jest podzielna przez żadną z badanych wartości")
```

## Zadanie 3: Pętla for
Napisz pętle, która iterować będzie po elementach listy *napisy*

Na podstawie jej treści zacznij budować łańcuch znaków (*string*). 

Podejmij następujące działania:
* Jeżeli wyraz w liście ma 4 lub więcej liter weź pierwsze dwa jego znaki
* Jeżeli wyraz ma równo 2 lub 3 litery weź 1 znak.
* Jeżeli wyraz ma 1 literę dodaj spację.

Wyświetl ukończony obiekt string - możesz dodać na końcu znak zapytania dla utrzymania porządku.  

```python
napisy = ["Polska", "z", "covid-19", "i", "torba", "a", "rak","obronił", "ile", "ela"]

# Pusty lańcuch
wynik = ""

# Dodawanie kolejnych elementów w pętli
for element in napisy:
  if len(element) >= 4:
    wynik += element[0:2]
  if len(element) == 2 or len(element) == 3:
    wynik += element[0]
  if len(element) == 1:
    wynik += " "

print(wynik, "?")
```

##Zadanie 4: Set albo Lista
Sprawdź w pętli, które elemnty z listy *druzyny_do_sprawdzenia* znajduja się w zbiorze *serie_a*

Dla każdego elementu wydrukuj: 
* Nazwa drużyny jest w Serie A - jeżeli drużyna znajduje się w zbiorze
* Nazwa drużyny spadł do Serie B - jeżeli drużyny nie ma w zbiorze. 



```python
druzyny_do_sprawdzenia = ["Benevento", "Empoli", "Sassuolo", "AC Parma"]

serie_a = {"Atalanta", "Bologna", "Cagliari", "Empoli", "Fiorentina", "Genoa", "Hellas Verona", 
           "Internazionale", "Juventus", "Lazio", "AC Milan", "Napoli", "Roma",  "Salernitana",
           "Sampdoria", "Sassuolo", "Spezia", "Torino", "Udinese", "Venezia" }

for element in druzyny_do_sprawdzenia:
  if element in serie_a:
    print(element, " jest w Serie A")
  else:
    print(element, " spadła do Serie B")
```

##Zadanie 5: Funkcje
Napisz funkcje, która liczy silnie. 

Wybierz jedno spośród 2 rozwiązań:

1. Iteracyjnie - zrobić w pętli iloraz od 1 do danego argumentu. 
2. Za pomocą rekurencji. 

W przypadku rozwiązania rekurencyjnego do napisania są takie warunki:

Jeżeli argument funkcji:
* jest większy od 1 zwróć wartość argumentu * wywołanie funkcji z argumentem pomniejszonym o 1
* To 1 zwróć jedynkę. 



```python
n = 6

# Rozwiazanie iteracyjne
wynik = 1
for i in range (1,n+1):
  wynik *= i
print(wynik)

# Rozwiazanie rekurencyjne
def silnia(n):
  if n == 1:
    return 1
  else:
    return n*silnia(n-1)

print(silnia(n))
```

#Zadanie 6: Metody String
Poniżej przedstawiam fragment utworu VirusSLD - kompletnie nie obchodzi nas jego polityczny wymiar, robimy ćwiczenia językowe.

1. Użyjmy metody *replace*, aby wyeliminowac bluzgi. 
2. Sprowadźmy tekst do małych liter. 
3. Wykorzystajmy metodę *split*, aby rozdzielić frazy. Niech separatorem będzie "jak ja"

```python
tekst = "Jak ja was kurwy nienawidzę I jak ja wami kurwy gardzę Jak ja się za was kurwy wstydzę Gdy za granicę czasem zajrzę"

# Krok 1
tekst = tekst.replace("kurwy","")
print("Tekst po usunięciu bluzgów:", tekst)

# Krok 2
tekst = tekst.lower()
print("Tekst bez wielkich liter:", tekst)

# Krok 2
listaWersow = tekst.split(sep="jak ja")
print(listaWersow)
```

## Zadanie 7: Funkcja / Lambda
Napisz funkcję albo wyrażenie lambda, które jako argument bierze 6 znakowy łańcuch znaków (*string*) - będzie udawać 6 pierwszych liczb z rejestru PESEL osób urodzonych przed 2000 np. 930101. 

Rozbijmy to tak, aby zaprezentowac datę urodzenia tj. 01 01 1993 

Czyli: 
2 ostatnie liczby > spacja > 2 środkowe liczby > spacja > 19 > 2 pierwsze liczby.

```python
napis = "930101"

f = lambda x: x[-2] + x[-1] + " " + x[len(x) - 4: len(x)-2] + " 19" + x[0:2]  
print(f(napis))
```
