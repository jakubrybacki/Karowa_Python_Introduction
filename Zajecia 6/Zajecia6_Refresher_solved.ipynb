### Zadanie 1
Napisz pętle, która zamieni polskie znaki w tekście na odpowiedniki zapisane w słowniku: 

**Podpowiedź**: Wykorzystaj metodę string *replace* oraz pętle *for*

Inspiracja: [Wikipedia](https://pl.wikipedia.org/wiki/Za%C5%BC%C3%B3%C5%82%C4%87_g%C4%99%C5%9Bl%C4%85_ja%C5%BA%C5%84)

````python
Tekst_do_przerobienia = "Zażółć gęślą jaźń"
slownik_znaki = { 
                 "ą": "a", 
                 "ć": "c",
                 "ę": "e", 
                 "ł": "l",
                 "ń": "n", 
                 "ó": "o",
                 "ś": "s",
                 "ż": "z",
                 "ź": "z"
                 }

#Trudność zadania polega na wybraniu odpowiedniej pętli - najlepiej moim zdaniem wykorzystać dwuelementową na kolekcji items
for element_klucz, element_wartosc in slownik_znaki.items():
  # W tym momencie przerabiamy w tekscie znaki z klucza (polski znak po lewej stronie od dwukropka w słowniku) na wartośc (odpowiednia litera bez ogonka - w słowniku po dwukropku)
  Tekst_do_przerobienia = Tekst_do_przerobienia.replace(element_klucz, element_wartosc)

print(Tekst_do_przerobienia)
````

### Zadanie 2
Napisz pętle, która:

* W momencie gdy wartość słownika zawiera tekst Wydrukuje 3 pierwsze litery z tekstu zmienione na wielkie znaki (uppercase). Pomocna będzie metoda *upper* 
* W momencie gdy wartość słownika zawiera liczbę wydrukuje nazwę klucza oraz tą liczbę.

Do sprawdzenia typu przydatna jest metoda *isinstance(zmienna, typ)*. Przykład jej zastosowania znajduje się poniżej: 



````python
text = "To jest text"
if isinstance(text, float):
  print(text + 1)
if isinstance(text, str):
  print(text)

liczba = "1.0"
if isinstance(liczba, float):
  print(liczba, " ma częśc dziesiętną")
elif isinstance(liczba, int):
  print(liczba, " nie ma w sobie ułamka")
````

Gotowy słownik i pole na kod poniżej:

````python
slownik_chaos = {
                 "gol1": 1,
                 "druzyna1": "Chelsea",
                 "gol2" : 2,
                 "team2": "Leicester"
                 }

# W pętli iterujemy po wszystkich elementach słownika
for element_klucz, element_wartosc in slownik_chaos.items():
  # Ten warunek sprawdza czy wartość jest typem tekstowym
  if isinstance(element_wartosc, str):
    # Jeżeli tak bierze pierwsze 3 znaki i powiększa je
    print(element_wartosc[0:3].upper())
  
  # W tym miejscu sprawdzany jest typ liczbowy
  if isinstance(element_wartosc, int):
    print(element_klucz, element_wartosc, sep = "\t")
````

### Zadanie 3 
Napisz program który policzy częstotliwości każdego wyrazu. Zwróć uwagę na fakt, że podczas przeglądania listy trzeba uwzględnić dwa warunki:
1. Elementu nie ma w słowniku i trzeba go dodać.
2. Element wystąpił wcześniej i trzeba odpowiadającą mu wartość słownika podnieść o 1. 

**Podpowiedź**: Istnienie klucza można sprawdzić podobnie jak w przypadku zbiorów. Taki kod:


```
# if "klucz" in dictionary
```

zwróci pożadany wynik.



````python
lista_wyrazow = "ala ma kota. kota ma ala. ala ma psa".split()
print("Wydrukujmy listę wyrazów:")
print(lista_wyrazow)
slownik_wyrazow = {}

# Pętla iteruje po wszystkich wyrazach z listy
for element in lista_wyrazow:

  # Każdy wyraz jest sprawdzany pod kątem wystapienia w słowniku
  if element in slownik_wyrazow:
     # Jeżeli element był - liczba jego wystąpień jest powiększana o 1 
     slownik_wyrazow[element] += 1
  else:
    # Jeżeli nie - ta linijka inicjuje nowy element z wartością 1
    slownik_wyrazow[element] = 1

print("")
print("Finalny słownik prezentuje się następująco:")
print(slownik_wyrazow)
````

Stwórz słownik *finalne_zapytanie* na podstawie dwóch słowników *podstawowe_zapytanie* i *szczegolowa_kwerenda*.

````python
from urllib.parse import urlencode

podstawowe_zapytanie = {  
    'page': 1,
    'limit': 100
    }

szczegolowa_kwerenda = {
    "filter[okres-rok]": 2021,
    "filter[okres-okres]": 2,
    "filter[jednostka-gt]": 0
    }

#Do przerobienia był wiersz poniżej - konieczne było rozpakowanie obydwu słowników przy pomocy podwójnego znaku mnożenia
finalne_zapytanie = dict(**podstawowe_zapytanie, **szczegolowa_kwerenda)

final_url = "https://www.test.com/search" + "?" + urlencode(finalne_zapytanie)
print(final_url)
````

### Zadanie 5
Wykorzystaj metode *type*, aby sprawdzić 

* Typ kolekcji *dict.items()*, *dict.keys()* oraz *dict.values()* dla słownika poniżej. 
* Typy elementów ze słownika w zależnosci od tego jaką kolekcje iterujemy w pętli.


````python
a = {"a": 1, "b":2}

# Tutaj sprawdzamy klasy z pierwszego punktu
print(type(a.items()), type(a.keys()), type(a.values()))

# Tutaj iterujemy w pętli po każdej z kolekcji z poprzedniego punktu:
print()
print("Typy - kolekcja elementy:")
for element in a.items():
  # T
  print(type(element), element)

print()
print("Typy - kolekcja kluczy:")
for element in a.keys():
  print(type(element), element)

print()
print("Typy - kolekcja wartosci:")
for element in a.values():
  print(type(element), element)
````
