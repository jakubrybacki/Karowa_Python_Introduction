Pętla zakończyć ma się gdy suma przekroczy 30. 

Wydrukuj: 
* sumę elementów
* indeks przy którym zakończyła się pętla.

```python
suma_elementy = 0
i = 0

while suma_elementy <= 30:

  #Zwróćcie uwagę na operator += - w obu przypadkach będzie on równoznaczny z kodami zapisanymi w komentarzach poniżej:
  #suma_elementy = suma_elementy +  losowa_lista[i]
  suma_elementy += losowa_lista[i]

  # Pamiętamy, że w odróżnieniu od pętli for tutaj musimy zdefiniować zmienną, która będzie 
  #   odpowiadała za przechodzenie między kolejnymi elementami tabeli:
  #i = i + 1
  i += 1

print("Suma elementów: ", suma_elementy)
print("Ostatni element który dodawała pętla: ", i)
```

### Zadanie 3
Poniższy kod wygeneruje listę złożoną z losowych 100 małych liter. Uruchom go:

```python
import random
import string

losowa_listaLiter = []
for i in range(0,100):
  n = random.choice(string.ascii_lowercase)
  losowa_listaLiter.append(n)

print(losowa_listaLiter)
```

Sprawdź ile liter w losowej liście to a, b, c bądź d. W warunku użyj odwołanie do zbioru liter:

**Rozwiązanie proste:**

Sprawdzimy ile liter występowało łącznie. Żeby sprawdzić czy, któraś litera występuje w zbiorze używamy następującej konstrukcji:




```python
zbior_liter = {'a', 'b', 'c', 'd'}

'a' in zbior_liter
# Otrzymujemy True w momencie gdy litera jest w zbiorze i False gdy jej nie ma. 
```

Teraz wystarczy tylko sprawdzić ten warunek w pętli dla wszystkich elementów listy.

Jeżeli warunek będzie spełniony zliczamy daną wartość. Tak działa kod poniżej: 

```python
zbior_liter = {'a', 'b', 'c', 'd'}

licznik = 0
for element in losowa_listaLiter:
  if element in zbior_liter:
    licznik +=1

print("Na liscie jest tyle liter a,b,c badź d: ", licznik) 
```

**Rozwiązanie złożone:**

Każdy problem można rozwiązać na kilka sposobów. Załóżmy, że chcemy jednak mieć wiedzę o każdej literze, a dodatkowo nie lubimy zliczania. 

Na poprzednich zajęciach nauczliśmy się filtrować listy, tak aby zawierały wyłącznie określone wartości. Umiemy też podać długość listy

Teraz wykorzystamy to dla naszych liter: 

```python
# Ta lista przefiltruje mi wszystkie litery a
listaA = [element for element in losowa_listaLiter if element == "a"]
print("Lista z literami 'a':", listaA, " ", sep = "\n")

# A ta komenda poda jej długość: 
dlugosc = len(listaA)
print("Długość listy z 'a':", dlugosc)
```

Mając taką wiedzę nie pozostaje mi nic innego jak zrobić pętlę, która przejdzie przez wszystkie elementy zbioru i zbierze wyniki:

```python
zbior_liter = {'a', 'b', 'c', 'd'}

# Tworzę pustą listę na wynik
wynik = []

# Tworzę pętle, która iteruje po zbiorze:  
for litera in zbior_liter:
  # Zmieniam warunek tak, aby filtrować listę na podstawie aktualnie 
  #   przeszukiwanej litery
  listaDoLitery = [element for element in losowa_listaLiter if element == litera]
  dlugosc = len(listaDoLitery)

  # Tworzę krotkę w której przechowywać będę literę oraz długośc listy
  krotka_licznosc = (litera, dlugosc)

  # Na koniec dodaje ją do listy wynikowej: 
  wynik.append(krotka_licznosc)

# Drukujemy wynik: 
print("Wynik:", wynik, sep = '\n')    
```

Zwróćcie uwagę, że kolejność wydruku będzie prawdopodobnie inna, niż ta wpisana początkowo. W set nie istnieje kolejnosć. 

Dlatego zamiast podawać same liczebności, tworzyliśmy krotki z parami: nazwa litery i dopiero jej liczbeność.

### Zadanie 4
Do zbioru *muzycy_zbior* dodaj róznymi metodami:
* Listę *polishMetal* 
* tekst *muzycy_30procent*

Użyj 2 różnych metod. 

Następnie ze zbioru usuń "Akcent" i "Sławomir".

```python
muzycy_zbior = {"Akcent", "Sławomir", "Kazik", "Kukiz"}
polishMetal = ["Vader", "Behemot", "Acid Drinkers", "Hunter"]
muzycy_30procent = "BFF"

# Dwie metody dodają elementy do zbioru:
#   update dodaje typy złożone - listy, sety etc. 
muzycy_zbior.update(polishMetal)

#   add dodaje typy proste - liczbę, łańcuch znaków (string)
muzycy_zbior.add(muzycy_30procent)
print("Zbiór po dodaniu wyrazów:", muzycy_zbior, " ", sep = "\n")

# Elementy najbezpieczniej usuwać metodą discard
muzycy_zbior.discard("Akcent")
muzycy_zbior.discard("Sławomir")
print("Końcowy zbiór:", muzycy_zbior, sep ="\n")
```

### Zadanie 5
Poniżej są teksty 2 piosenek AC/DC.

Wykonaj następujące polecenia: 
* Rozdziel teksty metodą split
* Przekonwertuj treść list do dwóch osobnych zbiorów.
* Znajdź sumę, iloczyn wspomnianych 2 zbiorów. Wygeneruje też unikatowe różnicę.

```python
ACDC_shootToTrill = "All you women who want a man of the street But don't know which way you wanna turn Just keep a-comin', and put your hand out to me 'Cause I'm the one who's gonna make you burn"
ACDC_TNT = "See me ride out of the sunset on your color TV screen Out for all that I can get If you know what I mean Women to the left of me And women to the right Ain't got no gun, ain't got no knife Don't you start no fight"

# W pierwszej kolejnosci rozbijamy teksty na listy i konwertujemy je do zbiorów.
#   Jak się uprzeć, całość działa w jednej linijce: 
ACDC_shootToTrill = set(ACDC_shootToTrill.split())
ACDC_TNT = set(ACDC_TNT.split())

# Teraz zostaje tylko logika:
print("Iloczyn zbiorów:")
iloczyn = ACDC_shootToTrill.intersection(ACDC_TNT)
print(iloczyn)
print('\n')

print("Suma zbiorów:")
suma = ACDC_shootToTrill.union(ACDC_TNT)
print(suma)
print('\n')

print("Elementy różne w dwóch zbiorach:")
roznice = ACDC_shootToTrill.symmetric_difference(ACDC_TNT)
print(roznice)
```
